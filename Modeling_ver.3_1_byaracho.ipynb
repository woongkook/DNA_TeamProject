### 1. Imputation

```python
df_train.isnull().sum()
```

```python
nan_cols = ['ind_empleado', 'pais_residencia', 'sexo', 'age', 'fecha_alta', 'ind_nuevo', 
            'indrel', 'indrel_1mes', 'tiprel_1mes', 'indresi', 'indext', 
            'canal_entrada', 'indfall', 'tipodom', 'cod_prov', 'ind_actividad_cliente', 'segmento']
nan_cols2 = ['antiguedad', 'renta']

for n in nan_cols:
    df_train[n] = df_train[n].fillna(df_train[n].mode()[0])
    
for n in nan_cols2:
    df_train[n] = df_train[n].transform(lambda x: x.fillna(x.mean()))
    
df_train['ind_nomina_ult1'] = df_train['ind_nomina_ult1'].transform(lambda x: x.fillna(0))
df_train['ind_nom_pens_ult1'] = df_train['ind_nom_pens_ult1'].transform(lambda x: x.fillna(0))

#or drop? -> ncodpers-test 사라지는지?
# train = train.dropna(axis=0)
```

### 3. Groupping

```python
df_train["age"] = pd.to_numeric(df_train["age"], errors="coerce")
df_test["age"] = pd.to_numeric(df_test["age"], errors="coerce")

age_group = [[0,19],[19,26],[26,36],[36,41],[41,47],[47,55],[55,60],[60,70],[70,80],[80,170]]

def create_age_group(df):  
    df['age_group'] = np.nan
    for i,age in enumerate(age_group):
        row_index = (df['age'] >= age[0]) & (df['age'] < age[1])
        df.loc[row_index,'age_group'] = i
        
create_age_group(df_train)
create_age_group(df_test)

df_train.drop('age', axis=1, inplace = True)
df_test.drop('age', axis=1, inplace = True)
```

```python
renta_group = [[0,50000],[50000,70000],[70000,100000],[100000,150000],[150000,200000],[200000,1000000],[1000000,29000000]]

def create_renta_group(df):  
    df['renta_group'] = np.nan
    for i,renta in enumerate(renta_group):
        row_index = (df['renta'] >= renta[0]) & (df['renta'] < renta[1])
        df.loc[row_index,'renta_group'] = i
        
create_renta_group(df_train)
create_renta_group(df_test)

df_train.drop('renta', axis=1, inplace = True)
df_test.drop('renta', axis=1, inplace = True)
```

### 7. Dummy

```python
dummy_col = ['sexo', 'canal_entrada', 'cod_prov', 'segmento', 'pais_residencia']
```

```python
limit = int(0.05 * len(df_train_m.index))

for col in dummy_col:
    if len(df_train_m[col].unique()) > 6:
        trainlist = df_train_m[col].value_counts()
        use_col = []
        for i,item in enumerate(trainlist):
            if item > limit:
                use_col.append(df_train_m[col].value_counts().index[i])       
        for item in df_train_m[col].unique(): 
            if item not in use_col:
                row_index = df_train_m[col] == item
                df_train_m.loc[row_index,col] = np.nan
        for item in df_test_m[col].unique(): 
            if item not in use_col:
                row_index = df_test_m[col] == item
                df_test_m.loc[row_index,col] = np.nan
```

```python
df_train_m = pd.get_dummies(df_train_m, prefix=dummy_col, columns = dummy_col)
df_test_m = pd.get_dummies(df_test_m, prefix=dummy_col, columns = dummy_col)
```

### Modeling

```python
from sklearn.ensemble import ExtraTreesClassifier
from collections import defaultdict
from sklearn.metrics import roc_auc_score
```

```python
models = {}
model_preds = {}
id_preds = defaultdict(list)
ids = df_test['ncodpers'].values

y_train = df_train_y
x_train = df_train_m

# train model here with june 2015 data
forest = ExtraTreesClassifier(max_depth=15, n_estimators=50, random_state=0)
forest.fit(x_train, y_train)
```

```python
importances = forest.feature_importances_
```

```python
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

print("Feature ranking:")
for f in range(df_train_m.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
# pridict model with the most recent data
x_train2 = df_test_m

for idx, c in enumerate(targets):  
    
    print(c)
    
    p_train = forest.predict_proba(x_train2)[idx][:,1]
#     p_train2 = forest.predict_proba(x_train)[idx][:,1]

    models[c] = forest
    model_preds[c] = p_train

    for id, p in zip(ids, p_train):
        id_preds[id].append(p)
```

```python
# check if customer already have each product or not. 
already_active = {}
for row in df_train_a.values:
    row = list(row)
    id = row.pop(0)
    active = [c[0] for c in zip(df_train_a.columns[1:], row) if c[1] > 0]
    already_active[id] = active
```
